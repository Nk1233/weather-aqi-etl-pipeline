```python
import os

import pandas as pd
import json
from sqlalchemy import create_engine, text
from dotenv import load_dotenv

load_dotenv()
```

```python
path1 = '../Json_data/Temp_City.json'
path2 = '../Json_data/Temp_Weather_records.json'
path3 = '../Json_data/Temp_Aqi_stations.json'
path4 = '../Json_data/Temp_Aqi_records.json'
path5 = '../Json_data/Temp_Forecast_records.json'
```

```python
city = pd.read_json(path1)
weather = pd.read_json(path2)
aqi_stations = pd.read_json(path3)
aqi_records = pd.read_json(path4)
forecast_records = pd.read_json(path5)
```

```python
City = pd.DataFrame(city)
Weather_records = pd.DataFrame(weather)
Aqi_stations = pd.DataFrame(aqi_stations)
Aqi_records = pd.DataFrame(aqi_records)
Forecast_records = pd.DataFrame(forecast_records)
```

```python
weather_url = os.getenv("DATABASE_URL")
```

```python
engine = create_engine(weather_url)
```

```python
City.to_sql('city', engine, if_exists='append', index=False)
```

```python
Weather_records.to_sql('weather_records', engine, if_exists='append', index=False)
```

```python
Aqi_stations.to_sql('aqi_stations', engine, if_exists='append', index=False)
```

```python
Aqi_records.to_sql('aqi_records', engine, if_exists='append', index=False)
```

```python
Forecast_records.to_sql('forecast_records', engine, if_exists='append', index=False)
```
