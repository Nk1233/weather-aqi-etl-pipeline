## Step 1: Import Required Libraries

- `requests`: For making HTTP calls to APIs
- `os`: For reading environment variables
- `dotenv`: For loading secrets from `.env` file

**Why `.env`?** Keeps API keys out of your code (security best practice)

````python
import requests
import os
from dotenv import load_dotenv

# Load environment variables from .env file
load_dotenv()
````

## Step 2: Load Your API Keys

Your `.env` file should contain:
```
OPENWEATHER_API_KEY=your_key_here
```

````python
# Fetch API keys from environment variables
api_key = os.getenv("OPENWEATHER_API_KEY")

# Verify key loaded (don't print the full key in production!)
print(f"API Key loaded: {api_key[:5]}..." if api_key else "ERROR: Key not found")
````

## Step 3: Build the API Request URL

OpenWeatherMap API endpoint for current weather:

**URL Structure:**
```
https://api.openweathermap.org/data/2.5/weather?q={city}&appid={key}&units=metric
```

**Parameters:**
- `q={city}` → City name
- `appid={key}` → Your API key
- `units=metric` → Get Celsius (not Kelvin)

````python
# Set the city we want to query
city = "Delhi"

# Construct the full API URL using f-string
url = f"http://api.openweathermap.org/data/2.5/weather?q={city}&appid={api_key}&units=metric"

print(f"Request URL: {url[:50]}...")  # Partial for safety
````

## Step 4: Make the API Request

`requests.get(url)` sends an HTTP GET request to the server.

**What happens:**
1. Your computer sends request to OpenWeatherMap server
2. Server processes it (finds weather for Delhi)
3. Server sends back JSON response
4. `response` object holds everything

````python
# Send GET request to the API
response = requests.get(url)

print(f"Response type: {type(response)}")
print(f"Status Code: {response.status_code}")
````

## Understanding Status Codes

| Code | Meaning |
|------|---------|
| 200 | OK - Success! |
| 404 | Not Found - City doesn't exist |
| 401 | Unauthorized - Bad API key |
| 429 | Too Many Requests - Rate limited |
| 500 | Server Error - Their problem |

**Always check `status_code == 200` before processing data!**

## Step 5: Parse the JSON Response

**What is JSON?**
JSON = JavaScript Object Notation
It's a way to structure data as nested key-value pairs.

In Python, JSON becomes:
- `{}` → Dictionary
- `[]` → List
- Nested structures → Nested dicts/lists

````python
# Convert JSON response to Python dictionary
data = response.json()

# Print raw data (messy but complete)
print(data)
````

## Step 6: Extract Useful Fields

Navigate the nested structure using keys:

```python
data['name']              # Top-level key
data['main']['temp']      # Nested: dict inside dict
data['weather'][0]['description']  # List index, then key
```

````python
# Extract specific fields from the nested structure
city_name = data['name']
temperature = data['main']['temp']      # Inside 'main' dict
humidity = data['main']['humidity']     # Also in 'main'
weather_desc = data['weather'][0]['description']  # First item in list

# Display formatted results
print(f"\n=== Weather for {city_name} ===")
print(f"Temperature: {temperature}°C")
print(f"Humidity: {humidity}%")
print(f"Condition: {weather_desc}")
````

## Practice: Try Different Cities

Change the `city` variable above and re-run:
- Try: "Mumbai", "London", "New York", "Tokyo"
- Watch how the JSON structure stays the same, only values change

````python
# Quick practice: Fetch another city
city2 = "London"
url2 = f"http://api.openweathermap.org/data/2.5/weather?q={city2}&appid={api_key}&units=metric"
resp2 = requests.get(url2)

if resp2.status_code == 200:
    d2 = resp2.json()
    print(f"{d2['name']}: {d2['main']['temp']}°C, {d2['weather'][0]['description']}")
else:
    print(f"Failed: {resp2.status_code}")
````
